```python
import json

import pandas as pd

data_path = "../data/srlm_ift.jsonl"

df = {
    "prompt_id": [],
    "prompt": [],
    "completion": [],
}

with open(data_path, "r", encoding="utf-8") as f:
    for line in f:
        sample = json.loads(line)
        df["prompt"].append(sample["prompt"])
        df["prompt_id"].append(sample["prompt_id"])
        df["completion"].append(sample["completion"])

pd.DataFrame(df).head()
```
